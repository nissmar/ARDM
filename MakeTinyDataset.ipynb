```python
import matplotlib.pyplot as plt
import numpy as np
import os
import PIL.Image as Image
from tqdm import tqdm
```

# CelebA

```python
root2 = 'CelebA/'

imgs = [e for e in os.listdir(root2) if '.jpg' in e]
imgs.sort()
imgs = imgs[10000: 30000]
L=[]

def one_hotize(im, k=32):
    ''' input: [0,1] HxW image
        output: {0,1} HxWxK image'''
    im = np.floor(im/256.0*k)/k
    return im

```

```python
im = Image.open(root2+imgs[np.random.randint(len(imgs))])

print(im.size)
plt.imshow(im)
plt.show()
#im.save('base.jpeg')
im = im.convert('L')
im = im.resize((60,73))
im = im.resize((178,218))
im = one_hotize(np.array(im), 32)
plt.imshow(im, cmap='gray')
plt.show()
#plt.imsave('reduced.jpeg', im, cmap='gray')
```

```python
target='TinyCelebA_10k2/'
```
